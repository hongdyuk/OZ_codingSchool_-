[문제2]

Python의 실수형(float) 계산 시 오차가 발생하는 이유에 대해 간단히 설명해 주세요

### 실수형 계산 시 오차가 발생하는 이유!!

실수형 계산 시 오차가 발생하는 이유는 컴퓨터가 실수를 2진 부동소수점 형식으로 표현하기 때문입니다.

유한한 비트 수, 2진법 표현의 한계, 반올림 오차등의 이유로 표현 방식은 일부 실수를 정확하게 나타내지 못합니다. 

1. 유한한 비트 수 : 컴퓨터는 유한한 비트 수를 사용하여 실수를 표현합니다. 따라서 무한한 소수 자릿수를 가지는 실수를 정확히 저장할 수 없습니다.

***

2. 2진법 표현의 한계 : 10진법에서 정확히 표현되는 많은 소수들이 2진법으로는 무한 소수로 표현됩니다. 예를 들어, 0.1은 2진법으로 무한 소수가 되어 정확하게 표현할 수 없습니다.

***

3. 반올림 오차 : 위 두 가지 이유로 인해, 실수를 저장하거나 계산할 때 반올림이 발생하여 미세한 오차가 생깁니다. 이 오차는 연산을 반복하면 누적될 수 있습니다.

[문제3]

슬라이싱에 대해 간단히 설명해주세요

- 슬라이싱이란?

슬라이싱은 Python에서 시퀀스 객체에서 일부분을 추출하는 기법을 말합니다. 슬라이싱은 객체의 일정 범위를 지정하여 그 부분을 새로운 객체로 반환합니다.

- 슬라이싱 기본 문법

sequence[start:stop:step]
start: 추출을 시작할 위치의 인덱스. 기본값은 0입니다.
stop: 추출을 종료할 위치의 인덱스 (해당 인덱스는 포함되지 않음). 기본값은 시퀀스의 끝까지입니다.
step: 추출할 때 건너뛸 간격을 지정합니다. 기본값은 1입니다.

- 주의사항

인덱스를 벗어나더라도 오류가 발생하지 않고 최대 범위까지 슬라이싱됩니다.
슬라이싱은 원본 객체의 일부를 새로운 객체로 반환하므로, 원본 객체는 변경되지 않습니다.
슬라이싱은 다양한 시퀀스 객체에 대해 사용할 수 있으며, 매우 유연하고 강력한 기능을 제공합니다.
